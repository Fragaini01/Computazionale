```python
import pandas as pd
import numpy as np
from scipy import stats as st
from matplotlib import pyplot as plt
```

```python
#Fridrich Hansen
FH = pd.read_excel('/home/francesco/Codici/Computazionale/Dc1/Punto A/All_palaeotemps.xlsx' ,skiprows=3,  sheet_name= 'Data compilation' , header = None , usecols='I,M' , names = ['Age ky\nbefore 2015' , 'Anomaly'])
FH.dropna(inplace=True)
FH['YB 2015'] = FH['Age ky\nbefore 2015']*1000
FH.drop(['Age ky\nbefore 2015'] , axis = 1 , inplace=True)
FH.describe()
```

```python
ZH = pd.read_excel('/home/francesco/Codici/Computazionale/Dc1/Punto A/All_palaeotemps.xlsx' , skiprows=3 , names = ['Age ky\nbefore 2015' , 'Anomaly'], sheet_name= 'Data compilation' , header = None , usecols='P,T')
ZH.dropna(inplace=True)
ZH['YB 2015'] = ZH['Age ky\nbefore 2015']*1000
ZH.drop(['Age ky\nbefore 2015'] , axis = 1 , inplace=True)
ZH.describe()
```

```python
LH = pd.read_excel('/home/francesco/Codici/Computazionale/Dc1/Punto A/All_palaeotemps.xlsx' , skiprows=3 , names = ['Age ky\nbefore 2015' , 'Anomaly'], sheet_name= 'Data compilation' , header = None , usecols='W,AA')
LH.dropna(inplace=True)
LH['YB 2015'] = LH['Age ky\nbefore 2015']*1000
LH.drop(['Age ky\nbefore 2015'] , axis = 1 , inplace=True)
LH.describe()
```

```python
EPICA = pd.read_excel('/home/francesco/Codici/Computazionale/Dc1/Punto A/All_palaeotemps.xlsx' , skiprows=3 , names = ['Age ky\nbefore 2015' , 'Anomaly'], sheet_name= 'Data compilation' , header = None , usecols='AC,AE')
EPICA.dropna(inplace=True)
EPICA['YB 2015'] = EPICA['Age ky\nbefore 2015']*1000
EPICA.drop(['Age ky\nbefore 2015'] , axis = 1 , inplace=True)
EPICA.describe()
```

```python
plt.plot(EPICA['YB 2015'] , EPICA['Anomaly'] , label = 'EPICA')
plt.plot(ZH['YB 2015'] , ZH['Anomaly'] , label = 'ZH')
plt.plot(FH['YB 2015'] , FH['Anomaly'] , label = 'FH')
plt.plot(LH['YB 2015'] , LH['Anomaly'] , label = 'LH')
plt.legend()
```
