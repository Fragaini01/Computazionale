```python
temperature = pd.read_csv('/home/francesco/Codici/Computazionale/Dc1/Punto B.1/deutnat.txt' , header = 85 , sep = '\t')
temperature.drop(['Depth corrected' , 'deut'] , axis = 1 , inplace = True)
temperature.head()
```

```python
co2 = pd.read_csv('/home/francesco/Codici/Computazionale/Dc1/Punto B.1/co2nat.txt' , header = None , skiprows=155 , names = ['Year' , 'Co2'] , sep = '\t')
co2.columns
```
