```python
import snscrape
import snscrape.modules.twitter as sn
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
Corriere = pd.read_csv('/home/francesco/Codici/Computazionale/Dc1/Punto B2/Riscaldamento_Globale_Corriere.csv')
Repubblica = pd.read_csv('/home/francesco/Codici/Computazionale/Dc1/Punto B2/Riscaldamento_Globale_Repubblica.csv')
Ansa = pd.read_csv('/home/francesco/Codici/Computazionale/Dc1/Punto B2/Riscaldamento_Globale_Ansa.csv')

Ansa.Data = pd.to_datetime(Ansa.Data)
Repubblica.Data = pd.to_datetime(Repubblica.Data)
Corriere.Data = pd.to_datetime(Corriere.Data)

Ansa['Anno'] = Ansa.Data.dt.year
Repubblica['Anno'] = Repubblica.Data.dt.year
Corriere['Anno'] = Corriere.Data.dt.year

Ansa.describe()

totale = pd.merge(Corriere , Repubblica , how = 'outer')
totale = pd.merge(totale , Ansa , how = 'outer')
```

```python
plt.hist(Ansa.Data)
plt.show()
plt.hist(Corriere.Data)
plt.show()
plt.hist(Repubblica.Data)
plt.show()

```

```python
bar = totale.groupby(['Anno']).Data.count()

bar.plot.bar(label = 'tweet con "Riscaldamento Globale" da ansa corriere e repubblica')
plt.legend()
```
