```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

import scipy
```

```python
def gamma(alpha , xm) : 
    x_prima = np.linspace(0 , xm , 10 * xm)
    x_dopo = np.linspace(xm , 1000 , 10 *  (1000-xm))

    y_prima = x_prima
    y_dopo = (xm/x_dopo) ** alpha

    x = np.append(x_prima , x_dopo)
    y= np.append(y_prima , y_dopo)

    return x, y
```

```python
plt.loglog()
plt.xlabel("x")
plt.ylabel("gamma(x)")
for i in range(4) : 
    d = 0.5 + 0.5*i
    plt.plot(gamma(d,1)[0] , gamma(d,1)[1] , label = "alpha: " + str(d) + "\nvar: " + str(round(gamma(d, 1)[1].var() , 3)) + "\nMean: " +  str(round(gamma(d , 1)[1].mean() , 3)))
    plt.legend()
```
